## Get all unique genres

```python
genres = set()
for item in genres_df.to_dict(orient='records'):
  genre = item['genres']  
  if isinstance(genre, dict):
      if genre:  # Check if the dictionary is not empty
          genre = list(genre.keys())[0]
      else:
          continue  # Skip to the next iteration if empty
  elif isinstance(genre, list):
      if genre:  # Check if the list is not empty
          genres.update(genre) 
      else:
          continue  # Skip to the next iteration if empty 
  else:
      continue # Handle unexpected data types

  genres.update(genre.split(','))
```

```python
genres = {genre.strip() for genre in genres}
```

```python
list(genres)
```

## Grouping the book id and genre according to the genres related to a book

```python
books_with_genres = book_df.merge(genres_df, on='book_id', how='left')
```

```python
books_with_genres = books_with_genres[['book_id', 'genres']]
```

```python
books_with_genres.head()
```

```python
def process_genres(row):
    genres_dict = row['genres']
    all_genres = []
    for genre_string in genres_dict.keys():
        all_genres.extend(genre_string.split(','))
    return [genre.strip() for genre in all_genres]
```

```python
books_with_genres['genres'] = books_with_genres.apply(process_genres, axis=1)
```

# Inserting data into the neo4j DB

```python
from neo4j import GraphDatabase

URI = "bolt://localhost:7689"
AUTH = ("neo4j", "12345678")
```

# Define database constraint

```python
def create_book_constraint(tx):
    queries = [
        "CREATE CONSTRAINT unique_book IF NOT EXISTS FOR (b:Book) REQUIRE (b.book_id, b.title) IS UNIQUE",
        "CREATE CONSTRAINT book_id_required IF NOT EXISTS FOR (b:Book) REQUIRE b.book_id IS NOT NULL",
        "CREATE CONSTRAINT book_title_required IF NOT EXISTS FOR (b:Book) REQUIRE b.title IS NOT NULL"
    ]
    for query in queries:
        tx.run(query)


def create_author_constraint(tx):
    queries = [
        "CREATE CONSTRAINT unique_author IF NOT EXISTS FOR (a:Author) REQUIRE a.author_id IS UNIQUE",
        "CREATE CONSTRAINT author_id_required IF NOT EXISTS FOR (a:Author) REQUIRE a.author_id IS NOT NULL"
    ]
    for query in queries:
        tx.run(query)


def create_genre_constraint(tx):
    queries = [
        "CREATE CONSTRAINT unique_genre IF NOT EXISTS FOR (g:Genre) REQUIRE g.name IS UNIQUE",
        "CREATE CONSTRAINT genre_name_required IF NOT EXISTS FOR (g:Genre) REQUIRE g.name IS NOT NULL"
    ]
    for query in queries:
        tx.run(query)


def create_user_constraint(tx):
    queries = [
        "CREATE CONSTRAINT unique_user IF NOT EXISTS FOR (u:User) REQUIRE u.user_id IS UNIQUE",
        "CREATE CONSTRAINT user_id_required IF NOT EXISTS FOR (u:User) REQUIRE u.user_id IS NOT NULL"
    ]
    for query in queries:
        tx.run(query)

```

# Execute constraint query

```python
with GraphDatabase.driver(URI, auth=AUTH) as driver:
    driver.verify_connectivity()

    with driver.session() as session:
        session.execute_write(create_book_constraint)
        session.execute_write(create_author_constraint)
        session.execute_write(create_genre_constraint)
        session.execute_write(create_user_constraint)
```

# View all constraints

```python
def view_all_constraints(driver):
    with driver.session() as session:
        result = session.run("SHOW CONSTRAINTS")
        constraints = [record for record in result]
        for constraint in constraints:
            print(dict(constraint))  # Print as a dictionary for readability
```

```python
with GraphDatabase.driver(URI, auth=AUTH) as driver:
    driver.verify_connectivity()

    view_all_constraints(driver)
```

# Create nodes

```python
def create_book(tx, book_id, title, isbn, description, image_url):
    tx.run(
        "CREATE (:Book {book_id: $book_id, title: $title, isbn: $isbn, description: $description, image_url: $image_url})",
        book_id=book_id, title=title, isbn=isbn, description=description, image_url=image_url)


def create_author(tx, author_id, name):
    tx.run(
        "CREATE (:Author {author_id: $author_id, name: $name})", author_id=author_id, name=name)


def create_genre(tx, genre_name: str):
    tx.run("CREATE (:Genre {name: $genre_name})", genre_name=genre_name)


def create_user(tx, u_id):
    tx.run("CREATE (:User {user_id: $u_id})", u_id=u_id)
```

# Create book written author relations

```python
def create_book_written_by_author_relationship(tx, book_id, author_id):
    query = """
        MATCH (b:Book), (a:Author)
        WHERE b.book_id = $book_id AND a.author_id = $author_id 
        MERGE (a)-[:WRITTEN_BY]->(b)
        """
    tx.run(query, book_id=book_id, author_id=author_id)
```

# Create book has genre relation

```python
def create_book_genre_relationship(tx, book_id, genre_name):
    query = """
        MATCH (b:Book), (g:Genre)
        WHERE b.book_id = $book_id AND g.name = $genre_name
        MERGE (b)-[:HAS_GENRE]->(g)
        """
    tx.run(query, book_id=book_id, genre_name=genre_name)
```

# Insert read by relation

```python
def create_user_book_relationship(tx, book_id, user_id, rating, review):
    query = """
        MATCH (b:Book), (u:User)
        WHERE b.book_id = $book_id AND u.user_id = $user_id
        MERGE (u)-[r:READING {rating: $rating, review: $review}]->(b)
        """
    tx.run(query, book_id=book_id, user_id=user_id, rating=rating, review=review)
```
